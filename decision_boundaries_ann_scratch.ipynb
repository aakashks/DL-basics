```python
from math import exp

def act(x, k=3):
    return 1 / (1 + exp(-k*x))

def full_net(x, y):
    z1 = x - 1
    z2 = - x - 2*y + 7
    z3 =  - x + 2*y - 1
    
    a1 = act(z1)
    a2 = act(z2)
    a3 = act(z3)
    
    net = - a1 - a2 - a3 + 1
    return act(net)
```

```python
# plot the function on 2d grid
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 20, 1000)
y = np.linspace(-10, 20, 1000)
xx, yy = np.meshgrid(x, y)
zz = np.zeros_like(xx)
for i in range(len(x)):
    for j in range(len(y)):
        zz[i,j] = full_net(xx[i,j], yy[i,j])
plt.imshow(zz, extent=[-10,20,-10,20], origin='lower', cmap='gray')
plt.show()
```
